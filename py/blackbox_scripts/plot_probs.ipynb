```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("bodyfat.csv").drop("Density", axis=1)
data = data - data.mean()
y = data["Bodyfat"].values
# centre the response to adjust for no intercept
y = y - np.mean(y)

x = data.drop("Bodyfat", axis=1)

N = x.shape[0]
D = x.shape[1]
```

```python
colnames = list(x.columns) + ["log-sigma"]
```

```python
def plot_probs(probs, colnames):

    fig, axes = plt.subplots(nrows=3, ncols=5, figsize=(20, 7))

    for i, ax in enumerate(axes.reshape(-1)):
        if i < D+1:
            ax.hist(probs[i], bins=10)
            ax.set_title(colnames[i])
        else:
            ax.set_visible(False)

    plt.tight_layout()
    plt.show()
```

```python
klvi_probs_mc = np.genfromtxt("./data/klvi_probs_mc.csv", delimiter=',')
klvi_probs_sis = np.genfromtxt("./data/klvi_probs_sis.csv", delimiter=',')
klvi_probs_psis = np.genfromtxt("./data/klvi_probs_psis.csv", delimiter=',')

chivi_probs_mc = np.genfromtxt("./data/chivi_probs_mc.csv", delimiter=',')
chivi_probs_sis = np.genfromtxt("./data/chivi_probs_sis.csv", delimiter=',')
chivi_probs_psis = np.genfromtxt("./data/chivi_probs_psis.csv", delimiter=',')
```

# KLVI

```python
plot_probs(klvi_probs_mc, colnames)
```

```python
plot_probs(klvi_probs_sis, colnames)
```

```python
plot_probs(klvi_probs_psis, colnames)
```

# CHIVI
